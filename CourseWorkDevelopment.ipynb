```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")
df
```

```python
df['Burn Rate']
```

```python
df.isnull().sum()
```

```python
# filling the null value with mean value of their particular columns 
df['Resource Allocation'].fillna(df['Resource Allocation'].mean(), inplace=True)
df['Mental Fatigue Score'].fillna(df['Mental Fatigue Score'].mean(), inplace=True)
df['Burn Rate'].fillna(df['Burn Rate'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(['Employee ID', 'Date of Joining'], axis=1, inplace=True)
```

```python
df.to_csv(r'S:\coursework\AI\courworkDevelopment\cleaned_dataset.csv', index=False)
```

```python
import matplotlib.pyplot as plt

gender_counts = df['Gender'].value_counts()

plt.figure()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
company_counts = df['Company Type'].value_counts()

plt.figure()
plt.pie(company_counts, labels=company_counts.index, autopct='%1.1f%%')
plt.title('Company Type')
plt.show()
```

```python
wfh_counts = df['WFH Setup Available'].value_counts()

plt.figure()
plt.pie(wfh_counts, labels=wfh_counts.index, autopct='%1.1f%%')
plt.title('WFH Setup Availability')
plt.show()
```
